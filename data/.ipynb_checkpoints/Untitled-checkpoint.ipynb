```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
import glob
from denoising.utils import *
import pandas
```

```python
from scipy.optimize import curve_fit

def sq_r_mse(x, x_hat):
    return np.sqrt(np.mean((x-x_hat)**2))

def stretched_decay(x, beta1, gamma, g_inf, alpha):
    return beta1 * np.exp(-2*(gamma * x)**alpha) + g_inf

def fit_stretched_exponent(x, y, y_er):
        
        
    param_bounds=([0,0,0.97,0.7],[2,0.6,1.2,2.2])
    
    N = len(y)
    
    alpha = 1
    g_inf = 1

    if y[0] < 1:
        g_0 = 0
    else:
        g_0 = y[0]-1
        
    y_fit = np.ones(N) * np.mean(y)
    R2_min = sq_r_mse(y, y_fit)
#     print(R2_min)
    
    res_vals = [0, 0, np.mean(y), 0]
    res_cov = np.zeros((4,4))
    
    for j in np.linspace(1,100, 5):
        gamma = j/N
        try: 
            best_vals, covar = curve_fit(stretched_decay, x, y, maxfev = 10000,
                                         sigma = y_er, absolute_sigma = True,
                                         p0 = [g_0, gamma, g_inf, 1],
                                         bounds = param_bounds)
            y_fit  = stretched_decay(x, *best_vals)
            R2 = sq_r_mse(y, y_fit)
#             print(R2, best_vals)
            if R2 < R2_min:
                R2_min = R2
                res_vals = best_vals
                res_cov  = covar
                break
            
        except:
            pass
#             print('cannot fit')

    
    return res_vals

def get_params(d):
    g, e = get_one_time_from_two_time(d)
    times = np.arange(len(g)-1) + 1
    return fit_stretched_exponent( times, g[1:], e[1:])
```

```python
datasets = []
for dataset in ['train', 'validation', 'test']:
    files = glob.glob(f'{dataset}_data/raw_g2/*')
# n_rois = []
# n_frames = []
    df = []
    for file in files:
        uid = file.split('_')[-1][2:]
        with open(file, 'rb') as openfile:
            x = pickle.load(openfile)
        for roi in range(x.shape[2]):
            params = get_params(x[:, :, roi])
            for k in range(x.shape[0]//25):
                new_record = Record(uid, roi, *params)
                df.append(new_record)
        xx = x[::2, ::2, :].copy()
        for roi in range(xx.shape[2]):
            params = get_params(xx[:, :, roi])
            for k in range(xx.shape[0]//25):
                new_record = Record(uid, roi, *params)
                df.append(new_record)
        n_rois.append(x.shape[2])
        n_frames.append(x.shape[0])
    datasets.append(pandas.DataFrame(df))
```

```python
df_train, df_val, df_test = datasets
```

```python
df_train = df_train[df_train.alpha !=0]
df_val = df_val[df_val.alpha !=0]
```

```python
fig, ax = plt.subplots(4,2, figsize = (10, 10), dpi = 300)

for i, par in enumerate(['gamma', 'beta', 'g_inf', 'alpha']):
    _ = ax[i,0].hist(df_train[par], bins = 100)
    _ = ax[i, 1].hist(df_val[par], bins = 100)
    
ax[0,0].set_title('training')
ax[0,1].set_title('validation')
ax[0,0].set_xlabel(r'$\Gamma [frames^{-1}]$')
ax[0,1].set_xlabel(r'$\Gamma [frames^{-1}]$')

ax[1,0].set_xlabel(r'$\beta$')
ax[1,1].set_xlabel(r'$\beta$')

ax[2,0].set_xlabel(r'$C_{\infty}$')
ax[2,1].set_xlabel(r'$C_{\infty}$')

ax[3,0].set_xlabel(r'$\alpha$')
ax[3,1].set_xlabel(r'$\alpha$')

plt.tight_layout()

```

```python
import torch
from collections import namedtuple
error = namedtuple('error', 'param perror')
val_errors = []
val_files = glob.glob('../output1/weight_decay_test_ld_20_200/validation_error*lr_1e-05_latent_space_200_batchsize_8_cv_10_20_k_1_1__0_wd*')
for f in val_files:
    if f[-3:] != 'png':
        wd = f.split('wd')[-1]
        values = torch.load(f)
        val_errors.append(error(wd, values))
        
plt.figure(dpi = 200)
for v in val_errors:
    plt.plot(v[1], label = f'weiht_decay = {v[0]}')
plt.legend()
plt.title('validation error for models with LD = 200')
```
